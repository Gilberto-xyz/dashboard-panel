### Procesamiento de datos menores

```python
# Rellenar los datos faltantes (NA) con 0 y crear la columna GDP per capita
co2_df = co2_df.fillna(0)
co2_df['gdp_per_capita'] = np.where(co2_df['population']!=0, co2_df['gdp']/co2_df['population'], 0)
```

```python
# Hacer interactivo el pipeline del dataframe
int_co2_df = co2_df.interactive()
```

### Emisión de CO2 a lo largo del tiempo por continente

```python
# Definir los paneles de widgets
year_slider = pn.widgets.IntSlider(name='Year slider', start=1750, end=2020, step=3, value=1850)
year_slider
```

```python
# Botones de radio para las medidas de CO2
yaxis_co2 = pn.widgets.RadioButtonGroup(
    name='Y axis',
    options=['co2', 'co2_per_capita'],
    button_type='success'
)
```

```python
continents = ['World', 'Asia', 'Europe', 'Africa', 'North America', 'South America', 'Antarctica']

co2_pipeline =(
    int_co2_df[
        (int_co2_df.year <= year_slider) & 
        (int_co2_df.country.isin(continents))
    ]
    .groupby(['country', 'year'])[yaxis_co2].mean()
    .to_frame()
    .reset_index()
    .sort_values(by='year')
    .reset_index(drop=True)
)
```
